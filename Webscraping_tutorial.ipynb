```python
#Step 1: FSC

from bs4 import BeautifulSoup
import requests
import pandas as pd


url = 'https://lhr.nu.edu.pk/faculty/'
source = requests.get(url)

soup = BeautifulSoup(source.text, 'html.parser')

faculty_section = soup.find('div', id='fsc')
faculty_members = faculty_section.find_all('div', class_='col-lg-3 col-md-4 col-sm-6 col-12')

faculty_data = []
No=0

for member in faculty_members:

        No=No+1

        idx=member.find('a',href=True)

        name = member.find('h5', class_='text-center')

        designation = member.find('p')

        email_tag=member.find('p',class_='mb-0 text-center')

        image_tag = member.find('img')
        image_url = image_tag['src'] if image_tag else 'N/A'

        hec_supervisor = 'HEC Approved PhD Supervisor' in member.text

        print("Sr",No)
        print("ID: ",idx['href'].split('/')[-1] if idx else 'N/A')
        print("Name", name.text.strip() if name else 'N/A')
        print("Designation", designation.text.strip() if designation else 'N/A')
        print( "HEC Approved PhD Supervisor", hec_supervisor)
        print("Email", email_tag.text.strip())
        print("Department :", "FAST School of Computing")
        print("ImageURL", image_url)
        print("\n")


        faculty_data.append({
            "Sr": No,
            "ID: ":idx['href'].split('/')[-1] if idx else 'N/A',
            "Name": name.text.strip() if name else 'N/A',
            "Designation": designation.text.strip() if designation else 'N/A',
            "HEC Approved PhD Supervisor": hec_supervisor,
            "Email": email_tag.text.strip(),
            "Department": "FAST School of Computing",
            "ImageURL": image_url
        })

df = pd.DataFrame(faculty_data)

df.to_csv('fsc.csv',index=False)

print("Data saved to fsc.csvðŸ˜ƒ")
```

```python
#Step no 2: Getting Extension and Latest Education

import pandas as pd
import requests
from bs4 import BeautifulSoup


fsc = pd.read_csv('fsc.csv')
ee = pd.read_csv('ee.csv')
cv = pd.read_csv('cv.csv')
fsm = pd.read_csv('fsm.csv')
ss = pd.read_csv('ss.csv')


def scrape(df,name):

    id_list = df['ID: '].tolist()

    results = []

    for id in id_list:

        url = f'http://lhr.nu.edu.pk/{name}/facultyProfile/{id}'

        response = requests.get(url)

        soup = BeautifulSoup(response.content, 'html.parser')

       #For Extension
        extension="N/A"
        phone_section = soup.find('p', class_='mt-0 text-center')
        phone_text = phone_section.find('span', class_='small').text
        extension = phone_text.split("Ext:")[1].strip() if "Ext:" in phone_text else "N/A"

        #For Highest Education
        education_latest="N/A"
        education_section = soup.find('div', class_='col-lg-8 col-md-6 col-sm-12 text-justify')
        if education_section:
            education_ul = education_section.find('ul')
            if education_ul:
                education_items = education_ul.find_all('li')
                if education_items:
                    education_latest = education_items[0].text.strip()



       #In this I am not printing results because they are too many i am instead direct appending them to files
        results.append({
                'ID: ': id,
                'Extension': extension,
                'Highest Education': education_latest
            })


    faculty_df = pd.DataFrame(results)

    faculty_df.to_csv(f'{name}_2.csv',index=False)
    print(f'Data saved to {name}_2.csv ðŸ¤§')



scrape(fsc, 'fsc')
scrape(ee, 'ee')
scrape(cv, 'cv')
scrape(fsm, 'fsm')
scrape(ss, 'ss')
```

```python
#Step 3 Mergeing all files

fsc = pd.read_csv('fsc.csv')
fsc_2 = pd.read_csv('fsc_2.csv')

ee = pd.read_csv('ee.csv')
ee_2 = pd.read_csv('ee_2.csv')

cv = pd.read_csv('cv.csv')
cv_2 = pd.read_csv('cv_2.csv')

fsm = pd.read_csv('fsm.csv')
fsm_2 = pd.read_csv('fsm_2.csv')

ss = pd.read_csv('ss.csv')
ss_2 = pd.read_csv('ss_2.csv')


fsc_merged = pd.merge(fsc, fsc_2, how='left', on='ID: ')
ee_merged = pd.merge(ee, ee_2, how='left', on='ID: ')
cv_merged = pd.merge(cv, cv_2, how='left', on='ID: ')
fsm_merged = pd.merge(fsm, fsm_2, how='left', on='ID: ')
ss_merged = pd.merge(ss, ss_2, how='left', on='ID: ')


all_faculty_data = pd.concat([fsc_merged, ee_merged, cv_merged, fsm_merged, ss_merged], ignore_index=True)

all_faculty_data.to_csv('fast_lhr_faculty.csv', index=False)

print("All Faculty Data Saved to fast_lhr_faculty.csv ðŸ˜©")
```
